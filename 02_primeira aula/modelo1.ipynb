```python
# FEATURES = CARACTERISDTICAS
# pelo longo 
# pernas curtas 
# faz au au 

# dados de treinamento : x 
porco1 = [1,1,0]
porco2 = [0,1,1]
porco3 = [0,1,0]

cachorro1 = [ 0,1,1]
cachorro2 = [ 1,1,1]
cachorro3 = [ 0,0,1]
```

```python
# 1-> porco 0-> cachorro 
treino_x = [porco1,porco2,porco3,cachorro1,cachorro2,cachorro3]
treino_y = [1,1,1,0,0,0]

# escoolha o algoritimo 
from sklearn.svm import LinearSVC
#  aqui ele e maiusculo por ser uma classe 

# terienar (FIT)
modelo = LinearSVC()
modelo.fit(treino_x,treino_y)
# y rotulos 
# x


```

```python
# animal misterioso
animal_misterioso = [1,0,0]
modelo.predict([animal_misterioso])
# a mudança e pq o porco esta naquelo  local do au au por isso nao e cachorro 
# sem espaços entre os itens 
```

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]
teste_x = [misterio1,misterio2,misterio3] # aqui o anil na posiçao 
previsoes = modelo.predict(teste_x)
```

```python
# classe correta - cachorro
teste_y = [0,1,1] # sim ou nao 
# FEATURES = CARACTERISDTICAS
# pelo longo 
# pernas curtas 
# faz au au 
```

```python
# 1-> porco 0-> cachorro 
#cachorro , porco , cachorro 

previsoes
```

```python
# Metricas
from sklearn.metrics import accuracy_score
taxa_de_acerto = accuracy_score(teste_y, previsoes)
```

```python
ta = taxa_de_acerto * 100
print(f"o modelo fez a previsão : cachorro, porco , cachorro ({ta:.2f} %)")
```
