for
O que é: Um laço que repete um bloco de código para cada item de uma sequência (lista, intervalo, etc).


Quando usar: Quando você sabe quantas vezes quer repetir, ou quer iterar sobre uma coleção (ex: de 1 a 100, uma lista de nomes).


Exemplo:

 for i in range(5):
    print(i)  # imprime 0,1,2,3,4



while
O que é: Um laço que repete o bloco enquanto uma condição for verdadeira.


Quando usar: Quando você não sabe exatamente quantas vezes vai repetir, mas quer continuar até uma condição mudar.


Exemplo:

 x = 0
while x < 5:
    print(x)
    x += 1



Resumo rápido:
Use for para repetir um número conhecido de vezes.


Use while para repetir até uma condição ser falsa.

Claro! Vamos detalhar o que significa cada parte dentro do `range(100, 0, -1)` e do código como um todo, explicando cada "espaço" ou termo usado.

---

```python
for numero in range(100, 0, -1):
    print(numero)
```

### Quebra do `range(100, 0, -1)`

* `range` — é uma função que cria uma **sequência de números**.
* `(100, 0, -1)` — são os **três argumentos** da função `range`:

  1. **100** — chamado de **valor inicial**. É o primeiro número da sequência.
  2. **0** — chamado de **valor final (exclusivo)**. A sequência vai até antes do 0, ou seja, termina no 1.
  3. **-1** — chamado de **passo** ou **incremento**. Diz para contar de trás para frente, diminuindo 1 a cada passo.

---

### Explicando o que significa cada argumento:

| Argumento | Nome               | Significado                          | Exemplo na sequência gerada    |
| --------- | ------------------ | ------------------------------------ | ------------------------------ |
| `100`     | Valor inicial      | Começa a contar por aqui             | Primeiro número é 100          |
| `0`       | Valor final        | Para antes desse número (não inclui) | Para antes de 0 (termina em 1) |
| `-1`      | Passo (incremento) | Como a contagem anda na sequência    | Decrementa 1 a cada número     |

---

### Exemplo da sequência gerada:

* Começa no **100**,
* Depois 99 (100 - 1),
* Depois 98 (99 - 1),
* ...
* Até chegar em 1 (porque para antes do 0).

---

### Significado da linha `for numero in ...`:

* **`for`**: "para cada..."
* **`numero`**: variável que vai armazenar o valor atual da sequência a cada passo do loop.
* **`in`**: indica de onde esses valores virão (a sequência gerada pelo `range`).

Então, “Para cada número na sequência do `range`, faça o que estiver indentado abaixo”.

---

### Linha `print(numero)`:

* Imprime o valor atual da variável `numero`.
